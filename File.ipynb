```python
import requests
import matplotlib.pyplot as plt
import numpy as np
from sklearn.linear_model import LinearRegression

# Obtener los datos de temperatura desde la API
temperature_url = 'https://iotbackendtemp.azurewebsites.net/api/temperature/'
temperature_response = requests.get(temperature_url)
temperature_data = temperature_response.json()

# Obtener los datos de humedad desde la API
humidity_url = 'https://iotbackendtemp.azurewebsites.net/api/humidity/'
humidity_response = requests.get(humidity_url)
humidity_data = humidity_response.json()

# Crear arrays para almacenar los datos de temperatura y humedad
temperature_values = []
humidity_values = []

# Extraer los valores de temperatura y humedad y almacenarlos en los arrays correspondientes
for data in temperature_data:
    temperature_values.append(data['value'])

for data in humidity_data:
    humidity_values.append(data['value'])

# Calcular el promedio de temperatura y humedad
temperature_avg = np.mean(temperature_values)
humidity_avg = np.mean(humidity_values)

# Agregar 100 datos cercanos al promedio al array de temperatura
temperature_values_extended = np.concatenate((temperature_values, np.random.normal(temperature_avg, 1, 100)))

# Agregar 100 datos cercanos al promedio al array de humedad
humidity_values_extended = np.concatenate((humidity_values, np.random.normal(humidity_avg, 1, 100)))

# Realizar la regresión lineal
X = np.array(temperature_values_extended).reshape(-1, 1)
y = np.array(humidity_values_extended)

model = LinearRegression()
model.fit(X, y)

# Realizar la predicción de humedad en función de la temperatura
temperature_range = np.linspace(min(temperature_values_extended), max(temperature_values_extended), 100)
humidity_predicted = model.predict(temperature_range.reshape(-1, 1))

# Crear el gráfico
plt.scatter(temperature_values_extended, humidity_values_extended, label='Datos')
plt.plot(temperature_range, humidity_predicted, color='red', label='Predicción')
plt.xlabel('Temperatura')
plt.ylabel('Humedad')
plt.legend()

# Mostrar el gráfico en Jupyter Notebook
plt.show()
```

```python
import requests
import matplotlib.pyplot as plt
import numpy as np
from sklearn.linear_model import LinearRegression

# Obtener los datos de temperatura desde la API
temperature_url = 'https://iotbackendtemp.azurewebsites.net/api/temperature/'
temperature_response = requests.get(temperature_url)
temperature_data = temperature_response.json()

# Obtener los datos de humedad desde la API
humidity_url = 'https://iotbackendtemp.azurewebsites.net/api/humidity/'
humidity_response = requests.get(humidity_url)
humidity_data = humidity_response.json()

# Crear arrays para almacenar los datos de temperatura y humedad
temperature_values = []
humidity_values = []

# Extraer los valores de temperatura y humedad y almacenarlos en los arrays correspondientes
for data in temperature_data:
    temperature_values.append(data['value'])

for data in humidity_data:
    humidity_values.append(data['value'])

# Calcular el coeficiente de correlación
correlation_coefficient = np.corrcoef(temperature_values, humidity_values)[0, 1]

# Realizar la regresión lineal
X = np.array(temperature_values).reshape(-1, 1)
y = np.array(humidity_values)

model = LinearRegression()
model.fit(X, y)

# Realizar la predicción de humedad en función de la temperatura
temperature_range = np.linspace(min(temperature_values), max(temperature_values), 100)
humidity_predicted = model.predict(temperature_range.reshape(-1, 1))

# Crear el gráfico de dispersión
plt.scatter(temperature_values, humidity_values, label='Datos')
plt.plot(temperature_range, humidity_predicted, color='red', label='Predicción')
plt.xlabel('Temperatura')
plt.ylabel('Humedad')
plt.legend()

# Mostrar el coeficiente de correlación en el gráfico
plt.text(plt.xlim()[0], plt.ylim()[1], f'Coeficiente de correlación: {correlation_coefficient:.2f}', ha='left', va='top')

# Mostrar el gráfico en Jupyter Notebook
plt.show()
```

```python
import requests
import matplotlib.pyplot as plt
import numpy as np

# Obtener los datos de temperatura desde la API
temperature_url = 'https://iotbackendtemp.azurewebsites.net/api/temperature/'
temperature_response = requests.get(temperature_url)
temperature_data = temperature_response.json()

# Obtener los datos de humedad desde la API
humidity_url = 'https://iotbackendtemp.azurewebsites.net/api/humidity/'
humidity_response = requests.get(humidity_url)
humidity_data = humidity_response.json()

# Crear arrays para almacenar los datos de temperatura y humedad
temperature_values = []
humidity_values = []

# Extraer los valores de temperatura y humedad y almacenarlos en los arrays correspondientes
for data in temperature_data:
    temperature_values.append(data['value'])

for data in humidity_data:
    humidity_values.append(data['value'])

# Crear el gráfico de barras
labels = ['Temperatura', 'Humedad']
values = [np.mean(temperature_values), np.mean(humidity_values)]

plt.bar(labels, values)
plt.xlabel('Variables')
plt.ylabel('Promedio')
plt.title('Promedio de Temperatura y Humedad')

# Mostrar el gráfico en Jupyter Notebook
plt.show()
```

```python
import requests
import matplotlib.pyplot as plt

# Obtener los datos de temperatura desde la API
temperature_url = 'https://iotbackendtemp.azurewebsites.net/api/temperature/'
temperature_response = requests.get(temperature_url)
temperature_data = temperature_response.json()

# Obtener los datos de humedad desde la API
humidity_url = 'https://iotbackendtemp.azurewebsites.net/api/humidity/'
humidity_response = requests.get(humidity_url)
humidity_data = humidity_response.json()

# Crear arrays para almacenar los datos de temperatura y humedad
temperature_values = []
humidity_values = []

# Extraer los valores de temperatura y humedad y almacenarlos en los arrays correspondientes
for data in temperature_data:
    temperature_values.append(data['value'])

for data in humidity_data:
    humidity_values.append(data['value'])

# Crear el histograma
fig, ax = plt.subplots(nrows=2, ncols=2, figsize=(10, 8))

ax[0, 0].hist(temperature_values, bins='auto', alpha=0.7)
ax[0, 0].set_xlabel('Temperatura')
ax[0, 0].set_ylabel('Frecuencia')
ax[0, 0].set_title('Histograma de Temperatura')

ax[0, 1].hist(humidity_values, bins='auto', alpha=0.7)
ax[0, 1].set_xlabel('Humedad')
ax[0, 1].set_ylabel('Frecuencia')
ax[0, 1].set_title('Histograma de Humedad')

ax[1, 0].scatter(temperature_values, humidity_values)
ax[1, 0].set_xlabel('Temperatura')
ax[1, 0].set_ylabel('Humedad')
ax[1, 0].set_title('Relación entre Temperatura y Humedad')

ax[1, 1].bar(['Temperatura', 'Humedad'], [len(temperature_values), len(humidity_values)])
ax[1, 1].set_xlabel('Variables')
ax[1, 1].set_ylabel('Cantidad de Lecturas')
ax[1, 1].set_title('Número de Lecturas')

plt.tight_layout()
plt.show()
```
